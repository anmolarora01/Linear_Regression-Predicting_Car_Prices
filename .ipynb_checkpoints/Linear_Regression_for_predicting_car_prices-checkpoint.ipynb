```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/CarPrice.csv')
```

```python
df.head()
```

```python
df.info()
```

The dataset does not contain any null values.

```python
#Checking unique car names
df.groupby(['CarName']).mean()['car_ID']
```

```python
#Split company name and car model from column "CarName" as the first word seems to b the company name
company_name = df['CarName'].str.split(" ", n=1, expand=True)
df["CarCompany"] = company_name[0]
df = df.drop(['CarName'],axis=1)
```

```python
#Checking unique company names in the dataset
df.groupby(['CarCompany']).mean()['car_ID']
```

```python
#Change the company name of "Nissan" to "nissan", "maxda" to "mazda", "porcshce" to "porsche", "toyouta" to "toyota", "vokswagen" and "vw" to "volkswagen"
df.loc[df['CarCompany'] == 'Nissan', 'CarCompany'] = 'nissan'
df.loc[df['CarCompany'] == 'maxda', 'CarCompany'] = 'mazda'
df.loc[df['CarCompany'] == 'porcshce', 'CarCompany'] = 'porsche'
df.loc[df['CarCompany'] == 'toyouta', 'CarCompany'] = 'toyota'
df.loc[df['CarCompany'] == 'vokswagen', 'CarCompany'] = 'volkswagen'
df.loc[df['CarCompany'] == 'vw', 'CarCompany'] = 'volkswagen'
```

```python
#Checking unique values of column "drivewheel"
df.groupby(['drivewheel']).mean()['car_ID']
```

```python
#Change 4wd to fwd
df.loc[df['drivewheel'] == '4wd', 'drivewheel'] = 'fwd'
```

```python
# List of variables to map

varlist =  ['doornumber', 'cylindernumber']

# Defining the map function
def number_map(x):
    return x.map({'one': 1, "two": 2, 'three':3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'twelve': 12})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(number_map)
```

```python
# create dummy variables for categorical variables and drop first type of every variable
dummyColumn = pd.get_dummies(df[['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem', 'CarCompany']], drop_first = True)
dummyColumn.head(20)
```

```python
dummyColumn.info()
```

```python
# add dummy columns to main dataframe
df = pd.concat([df, dummyColumn], axis=1)
df.head(20)
```

```python
# drop columns from which dummy variables were created
df = df.drop(['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem', 'CarCompany'],axis=1)
df.head()
```

```python
df.to_csv('CarsProcessedData.csv', sep=',')
```

# Splitting data and Building Model

```python
# creation of train and test dataset objects. (70-30 split)
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Applying scaler() to all the columns except 'dummy' variables
num_vars = ['symboling', 'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
# creating x_train and y_train for model
y_train = df_train.pop('price')
X_train = df_train
```

```python
#Build a linear model

import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.params
```

```python
print(lr_1.summary())
```

```python
#  Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = vif['VIF'].round(2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
